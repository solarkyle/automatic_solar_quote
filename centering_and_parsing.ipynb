```python
def pix_per_class(img):
    """returns unique dictionary with class numbers for keys and number of pixels per class as values"""
    unique, counts = np.unique(img, return_counts=True)
    unique = [int(i) for i in unique]
    act_classes = dict(zip(unique, counts))
    
    return act_classes
```

```python

for im in listdir('seg_data/images_and_masks'):

    image = Image.open(f'seg_data/images_and_masks/{im}')
    mask_name = im.replace(".jpg", ".png")
    mask = Image.open(f'seg_data/masks_gab_hip/{mask_name}')
    npimage = np.array(image)
    npmask = np.array(mask)
    original_mask = npmask.copy()
    original_image = npimage.copy()
    npmask[np.where(npmask == 19)] = 0
    npmask[np.where(npmask == 2)] = 0
    npmask[np.where(npmask != 0)] = 1
    kernel = np.ones((3,3),np.uint8)
    npmask = cv2.dilate(npmask,kernel,iterations = 1)
    contour, hier = cv2.findContours(npmask, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
    lst = []
    contour_indices = []
    for i in range(len(contour)):
        if hier[0][i][3] == -1:
            area = cv2.contourArea(contour[i])
            if area > 7000:
                coords = contour[i].reshape(-1, 2)
                lst.append(coords.mean(axis=0))
                contour_indices.append(i)

    arr = np.zeros((len(contour_indices),512,512))
    for k, e in enumerate(contour_indices):
        cv2.drawContours(arr[k], contour, e, 255, -1)

    for i,e in enumerate(lst):
        original_mask_copy = original_mask.copy()
        original_image_copy = original_image.copy()

        original_mask_copy[np.where(arr[i] == 0)] = 0
        original_image_copy[np.where(arr[i] == 0)] = 0

        subsums = 256 - lst[i]

        rolledmask1 = np.roll(original_mask_copy,int(subsums[1]),axis=0)
        rolledmask2 = np.roll(rolledmask1,int(subsums[0]),axis=1)

        rolledimage1 = np.roll(original_image_copy,int(subsums[1]),axis=0)
        rolledimage2 = np.roll(rolledimage1,int(subsums[0]),axis=1)

        new_mask = Image.fromarray(rolledmask2)
        new_mask.save(f"seg_data/singled_masks/{i}-{mask_name}")
        new_image = Image.fromarray(rolledimage2)
        new_image.save(f"seg_data/singled_images/{i}-{im}")
```
